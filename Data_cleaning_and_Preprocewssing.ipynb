```python
import pandas as pd
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df.head()
```

```python
df.isnull().sum() #missing values
```

```python
df = df.drop_duplicates() # Remove duplicates
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

```python
df["z_revenue"] = df["z_revenue"].fillna(df["z_revenue"].median())
```

```python
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
df.columns = df.columns.str.strip()
df["education"] = df["education"].str.strip()
```

```python
df['marital_status'] = df['marital_status'].replace({
    'Absurd': 'Absurd',
    'Yolo': 'Yolo',
    'Together': 'Together',
    'Divorced': 'Divorced'
})
```

```python
print(df['marital_status'].unique())
```

```python
Q1 = df["income"].quantile(0.25)
Q3 = df["income"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["income"] < lower) | (df["income"] > upper)]

print(outliers)
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_") # Rename columns
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True) # Convert date column
```

```python
df['education'] = df['education'].str.title()
df['marital_status'] = df['marital_status'].str.title() # Standardize text columns
```

```python
df['income'] = df['income'].fillna(df['income'].median()).astype(int) # Fix data types
```

```python
df.sort_values(by="id", ascending=False, inplace=True)
```
