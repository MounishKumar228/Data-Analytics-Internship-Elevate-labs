```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
df = pd.read_csv("Titanic Dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df["Survived"].value_counts())

print(df["Sex"].value_counts())

print(df["Pclass"].value_counts())

print(df["Embarked"].value_counts())
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes(include=['number']).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df.select_dtypes(include=['number']),
    hue="Survived"
)

plt.show()
```

```python
df.hist(figsize=(12,10))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Age"])

plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Fare"])

plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Age",
    y="Fare",
    hue="Survived",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Survived",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Sex",
    hue="Survived",
    data=df
)

plt.show()
```

```python
sns.countplot(
    x="Pclass",
    hue="Survived",
    data=df
)

plt.show()
```

```python
sns.histplot(
    df["Age"],
    bins=30,
    kde=True
)

plt.show()
```

```python
sns.histplot(
    df["Fare"],
    bins=30,
    kde=True
)

plt.show()
```
